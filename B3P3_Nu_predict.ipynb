```python
aac_file = "./Features/AAC_features.tsv"
ctdc_file = "./Features/CTDC_features.tsv"
ctdt_file = "./Features/CTDT_features.tsv"
ctdd_file = "./Features/CTDD_features.tsv"

df_aac = pd.read_csv(aac_file, sep='\t')
df_ctdc = pd.read_csv(ctdc_file, sep='\t')
df_ctdt = pd.read_csv(ctdt_file, sep='\t')
df_ctdd = pd.read_csv(ctdd_file, sep='\t')

label_column = '#'

features = pd.concat([df_aac[label_column], df_aac.iloc[:, 1:], df_ctdc.iloc[:, 1:], df_ctdt.iloc[:, 1:], df_ctdd.iloc[:, 1:]], axis=1)

output_file = "features.tsv"
features.to_csv(output_file, sep='\t', index=False)
features
```

```python
prep = features.drop(columns=['#'])
```

```python
prep
```

```python
features_sorter = pd.read_csv("./B3P3-Nu/prediction/Features_sorted.csv")
features_sorter.head()
```

```python
threshold_value = 0
filtered_features_0_sort = features_sorter[features_sorter['sorted'] > threshold_value]
filtered_features_0 = filtered_features_0_sort['Features'].astype(str).tolist()
print(len(filtered_features_0))
```

```python
predictor = prep[filtered_features_0]
predictor
```

```python
nusvc_model = joblib.load('./B3P3-Nu/prediction/nusvc_model.joblib')
```

```python
autoproba_nusvc = nusvc_model.predict_proba(predictor)
autopred_nusvc = nusvc_model.predict(predictor)
```

```python
Probability_column= autoproba_nusvc[:, 1]
Preditiction = pd.DataFrame({
    'Peptide number' : features["#"],
    'Probability': Probability_column,
    'Prediction': autopred_nusvc
})
```
